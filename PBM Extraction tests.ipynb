```python
import re
import math
import sys
```

```python
filename = "/opt/pbm/ornithorinque.pbm"
```

```python
with open(filename,'r') as f:
    while True:
        header = f.readline().strip()
        if header.startswith('#'):
           continue
        elif header == 'P1':
           assert False, 'ASCII format not supported'
        elif header == 'P4':
           break
        else:
           assert False, 'Bad format !'
    
    
    rows, cols = 0, 0
    while True:
        header = f.readline().strip()
        if header.startswith('#'):
           continue
        else:
           match = re.match('^(\d+) (\d+)$',header)
           if match is None:
               assert False, 'Bad size'
           else:
               cols, rows = match.groups()
               break
    assert (rows, cols) != (0, 0)
    colbytes = int(math.ceil(int(cols)/8.0))
    rows = int(rows)

    xL, xH = colbytes % 256, int(math.floor(colbytes / 256.0))
    yL, yH = rows % 256, int(math.floor(rows / 256.0))
    #xL, xH = 0,0
    #yL, yH = 0,0
    
    print("xL:%s xH:%s yL:%s yH:%s" %(xL, xH, yL, yH))
    
    cnt = 0
    print("{")
    for n in range(0,int(rows)):
        for m in range(0,colbytes):
            mychar = f.read(1)
            cnt += 1
            print'0x%02X,' % ord(mychar),
    print("}")
    print(cnt)
    
```
